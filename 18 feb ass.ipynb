Web API is an API as the name suggests, it can be accessed over the web using the HTTP protocol. 

API is general term which means to connect the two systems on the other hand web API is subset of API which means connecting with the server is held on web browser.

REST is a set of guidelines that offers flexible implementation, whereas SOAP is a protocol with specific requirements like XML messaging.

SOAP uses XML format that must be parsed to be read. It defines many standards that must be followed while developing the SOAP applications. So it is slow and consumes more bandwidth and resource.

soap Transports data in standard XML format whereas rest Generally transports data in JSON.

soap works with WSDL whereas rest works with GET, POST, PUT, DELETE
